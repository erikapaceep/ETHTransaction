```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bq-results-2015-08-20_2015-08-09.csv')
```

```python
len(df)
```

Select the transaction address from and to

```python
df_networkx = df[['from_address','to_address']]
transaction_address = list(zip(df_networkx.from_address,df_networkx.to_address))
```

```python
G = nx.MultiDiGraph()
G.add_edges_from(transaction_address)
nx.draw(G)
```

## Degree centrality

```python
degree_centrality=nx.degree_centrality(G)
```

## Betwness centrality

```python
between_centrality = nx.betweenness_centrality(G)
```
